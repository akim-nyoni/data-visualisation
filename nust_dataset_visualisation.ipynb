```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('nust_dataset_twitter_scrape.csv')
```

```python
data.head()
```

```python
retweet = data['retweetCount'].mean()
reply = data['replyCount'].mean()
like = data['likeCount'].mean()
quote = data['quoteCount'].mean()
```

```python
print("Retweet Count Mean is " + str(retweet) + "\nReply Count Mean is " + str(reply) + "\nLike Count Mean is " + str(like) + "\nQuote Count Mean is " + str(quote))
```

```python
labels = ['Retweet', 'Reply', 'Like', 'Quote']
sizes = [retweet, reply, like, quote]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

plt.axis('equal')

plt.title('Distribution of Mean or Average')
plt.savefig('dist_ave.png')
plt.show()
```

```python
retweetmax = data['retweetCount'].max()
replymax = data['replyCount'].max()
likemax = data['likeCount'].max()
quotemax = data['quoteCount'].max()
```

```python
print("Retweet Count Max is " + str(retweetmax) + "\nReply Count Max is " + str(replymax) + "\nLike Count Max is " + str(likemax) + "\nQuote Count Max is " + str(quotemax))
```

```python
labels = ['Retweet', 'Reply', 'Like', 'Quote']
sizes = [retweetmax, replymax, likemax, quotemax]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

plt.axis('equal')

plt.title('Distribution of Maximum')
plt.savefig('dist_max.png')
plt.show()
```

```python
data['createdAt'] = pd.to_datetime(data['createdAt'], format="%a %b %d %H:%M:%S %z %Y")
```

```python
data.head()
```

```python
data.sort_values('createdAt', inplace=True)
```

```python
plt.plot(data['createdAt'], data['retweetCount'])
plt.xlabel('Date')
plt.ylabel('Retweet Count')
plt.title('Retweet Count Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('line_retweet.png', bbox_inches='tight')
plt.show()
```

```python
plt.plot(data['createdAt'], data['replyCount'])
plt.xlabel('Date')
plt.ylabel('Reply Count')
plt.title('Reply Count Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('line_reply.png', bbox_inches='tight')
plt.show()
```

```python
plt.plot(data['createdAt'], data['likeCount'])
plt.xlabel('Date')
plt.ylabel('Like Count')
plt.title('Like Count Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('line_like.png', bbox_inches='tight')
plt.show()
```

```python
plt.plot(data['createdAt'], data['quoteCount'])
plt.xlabel('Date')
plt.ylabel('Quote Count')
plt.title('Quote Count Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('line_quote.png', bbox_inches='tight')
plt.show()
```

```python
data['RetweetClassification'] = data['isRetweet'].map({True: 'Posts with Retweet', False: 'Posts without Retweet'})
```

```python
retweet_classification_counts = data['RetweetClassification'].value_counts()
```

```python
sns.barplot(x=retweet_classification_counts.index, y=retweet_classification_counts.values)
plt.xlabel('Retweet Classification')
plt.ylabel('Count')
plt.title('Retweet Post Classification')
plt.savefig('bar_chart_retweet_class.png')
plt.show()
```

```python
data['QuoteClassification'] = data['isQuote'].map({True: 'Posts that are Quoted', False: 'Posts that are not Quote'})
```

```python
quote_classification_counts = data['QuoteClassification'].value_counts()
```

```python
sns.barplot(x=quote_classification_counts.index, y=quote_classification_counts.values)
plt.xlabel('Quote Classification')
plt.ylabel('Count')
plt.title('Quoted Post Classification')
plt.savefig('bar_chart_quote_class.png')
plt.show()
```
